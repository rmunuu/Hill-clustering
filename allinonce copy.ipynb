```python
grid = 400
s = 0.05

fig = plt.figure(figsize=(5, 5))

x = np.linspace(0, 1, grid+1)
y = np.linspace(0, 1, grid+1)
x, y = np.meshgrid(x, y)

z = np.zeros_like(x)
for point in data:
    z += gaussian(x, y, point[0], point[1], s)

ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis', edgecolor='none')

plt.show()
```
