```python
X1 = make_circles(factor=0.5, noise=0.05, n_samples=500)[0]
X2 = make_moons(n_samples=500, noise=0.05)[0]
X3 = make_blobs(n_samples=500, random_state=42, cluster_std=3)[0]
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(0, 1))
if len(X1.shape) == 1:
    X1 = X1.reshape(-1, 1)

scaler = MinMaxScaler(feature_range=(0, 1))
if len(X2.shape) == 1:
    X2 = X2.reshape(-1, 1)

scaler = MinMaxScaler(feature_range=(0, 1))
if len(X3.shape) == 1:
    X3 = X3.reshape(-1, 1)
```

```python
data = scaler.fit_transform(X1)
```

```python
level_bin_array, edge = simul_edge(data, (1, 1), 400, 100, sigma=0.07, hill_method=1, k=2)

edge_array = [[1 for _ in range(103)] for _ in range(103)]
for i, j in edge:
    edge_array[i][j] = 0
edge_array = np.array(edge_array)
data = np.array(data)
plt.imshow(~edge_array[::-1], cmap='binary', extent=[0, 1, 0, 1])
plt.scatter(data[:, 0], data[:, 1])
plt.title("sigma=0.05, k=2")
plt.show()
```
