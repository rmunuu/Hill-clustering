```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import (make_blobs, make_circles, make_moons)
from sklearn.preprocessing import MinMaxScaler
```

```python
data1 = make_circles(factor=0.5, random_state=42, noise=0.05, n_samples=500)[0]
data2 = make_moons(n_samples=500, random_state=42, noise=0.05)[0]
data3 = make_blobs(n_samples=500, random_state=42, cluster_std=2)[0]
scaler = MinMaxScaler(feature_range=(0, 1))
if len(data1.shape) == 1:
    data1 = data1.reshape(-1, 1)
data1 = scaler.fit_transform(data1)
if len(data2.shape) == 1:
    data2 = data2.reshape(-1, 1)
data2 = scaler.fit_transform(data2)
if len(data3.shape) == 1:
    data3 = data3.reshape(-1, 1)
data3 = scaler.fit_transform(data3)

plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.scatter(data1[:, 0], data1[:, 1], s=6)
plt.title('data1')

plt.subplot(1, 3, 2)
plt.scatter(data2[:, 0], data2[:, 1], s=6)
plt.title('data2')

plt.subplot(1, 3, 3)
plt.scatter(data3[:, 0], data3[:, 1], s=6)
plt.title('data3')

plt.show()
```

```python
X1 = make_circles(factor=0.5, random_state=42, noise=0.05, n_samples=500)
X2 = make_moons(n_samples=500, random_state=42, noise=0.05)
X3 = make_blobs(n_samples=500, random_state=42, cluster_std=2)

plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.scatter(X1[0][:, 0], X1[0][:, 1], s=6)
plt.title('data1')

plt.subplot(1, 3, 2)
plt.scatter(X2[0][:, 0], X2[0][:, 1], s=6)
plt.title('data2')

plt.subplot(1, 3, 3)
plt.scatter(X3[0][:, 0], X3[0][:, 1], s=6)
plt.title('data3')

plt.show()
# X1[0]
```

```python
from mpl_toolkits.mplot3d import Axes3D

X = X2[0]
sigma = 0.2

# Function to generate a Gaussian peak at each data point
def gaussian(x, y, x0, y0, sigma):
    return np.exp(-((x - x0)**2 + (y - y0)**2) / (2 * sigma**2))

# Meshgrid for evaluation
grid = 400
x = np.linspace(np.min(X[:, 0]) - 1, np.max(X[:, 0]) + 1, grid+1)
y = np.linspace(np.min(X[:, 1]) - 1, np.max(X[:, 1]) + 1, grid+1)
xm, ym = np.meshgrid(x, y)

ball = 10
ball_x = np.linspace(np.min(X[:, 0]) - 1, np.max(X[:, 0]) + 1, ball+1)
ball_y = np.linspace(np.min(X[:, 1]) - 1, np.max(X[:, 1]) + 1, ball+1)
ball_x, ball_y = np.meshgrid(ball_x, ball_y)


# Compute the sum of Gaussians at each grid point
z = np.zeros_like(xm)
for point in X:
    z += gaussian(xm, ym, point[0], point[1], sigma)

# ball_z = np.zeros_like(ball_x)
ball_z = np.array([[z[i][j] for i in np.arange(0, 400+400//ball, 400//ball)] for j in np.arange(0, 400+400//ball, 400//ball)])

# 3D plot
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(xm, ym, z, cmap='viridis', edgecolor='none')
ax.scatter3D(ball_x, ball_y, ball_z, c='r', s=20)
plt.axis("off")
plt.show()


print(ball_x[0])
print([x[i] for i in np.arange(0, 400+400//ball, 400//ball)])

print(ball_z[0])
print([z[i][0] for i in np.arange(0, 400+400//ball, 400//ball)])
```

```python
from mpl_toolkits.mplot3d import Axes3D

X = X2[0]

# Function to generate a Gaussian peak at each data point
def gaussian(x, y, x0, y0, sigma):
    return np.exp(-((x - x0)**2 + (y - y0)**2) / (2 * sigma**2))

# Meshgrid for evaluation
x = np.linspace(np.min(X[:, 0]) - 1, np.max(X[:, 0]) + 1, 400)
y = np.linspace(np.min(X[:, 1]) - 1, np.max(X[:, 1]) + 1, 400)
x, y = np.meshgrid(x, y)

def plot_graph(sigma):
    # Compute the sum of Gaussians at each grid point
    z = np.zeros_like(x)
    for point in X:
        z += gaussian(x, y, point[0], point[1], sigma)

    # 3D plot
    fig = plt.figure(figsize=(3, 2))
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(x, y, z, cmap='viridis', edgecolor='none')
    plt.axis("off")
    plt.show()

n = 10

sigma_array = np.linspace(0, 1, 10)
print(sigma_array)

for i in range(n):
    plot_graph(sigma_array[i])
```
